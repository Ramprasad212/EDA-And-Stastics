# Detect outliers using IQR

**GENERAL FORMULA FOR OUTLIERS DETECTION WITH IQR**

# Standard Formula to detect outliers


```python
df
```

```python
q3= df['height'].quantile(0.75)
q1= df['height'].quantile(0.25)
iqr= q3-q1
print(iqr)
ll = q1- 1.5 * (iqr)
ul = q3+ 1.5 * (iqr)
print(ll,ul)
```

```python
# check outliers
df[(df['height'] <=ll) | (df['height'] >=ul)]
```

```python
# check outliers
a= df.loc[(df['height'] <=ll) | (df['height'] >=ul),'height']
print(a)
l = list(a)
l
```

```python
for i in range(len(df)):
    if df.loc[i,'height'] in l:
        df.loc[i,'height'] = np.NAN
df
```

```python
df['height'].fillna(df['height'].mean())
```

```python
# check data without outliers
df[~((df['height'] <=ll) | (df['height'] >=ul))]
```

```python
IQR= Q3-Q1
```

# 2. Capping

```python
Q1 = df['height'].quantile(0.25)
Q3 = df['height'].quantile(0.75)
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
print(lower_limit,upper_limit)
```

```python
ll=max(min(df['height']),Q1 - 1.5*IQR)
ul=min(max(df['height']),Q3 + 1.5*IQR)
```

```python
ll,ul
```

```python
df
```

```python
import seaborn as sns
sns.boxplot(x= df['height'])
```

```python
df_cap = df.copy()
import numpy as np
df_cap['new'] = np.where(
    df_cap['height'] > ul,
    ul,
    np.where(
        df_cap['height'] <ll ,
        ll,
        df_cap['height']))
```

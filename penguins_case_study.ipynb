```python
import seaborn as sns
f = sns.load_dataset('flights')
f
```

```python
new= f.groupby(by='year').sum('passengers')
new.reset_index(inplace=True)
new
```

```python
sns.lineplot(data=new,x='year',y='passengers')
```

```python
f
```

```python
temp = f.groupby(by='month',observed=False)[['passengers']].sum()
temp.reset_index(inplace=True)
```

```python
temp
```

```python
sns.lineplot(data= temp,x='month',y='passengers')
```
